```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
img = cv2.imread('JGS_1.jpeg')

# Convert the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform histogram equalization to enhance the image
img_eq = cv2.equalizeHist(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY))
img_eq = cv2.cvtColor(img_eq, cv2.COLOR_GRAY2RGB)

# Plot the original image and the enhanced image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_eq)
plt.title('Enhanced Image')

plt.show()
```

### Keep the colors

```python
# Load the image
img = cv2.imread('JGS_1.jpeg')

# Convert the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split the image into its 3 color channels
b, g, r = cv2.split(img)

# Perform histogram equalization on each color channel
b_eq = cv2.equalizeHist(b)
g_eq = cv2.equalizeHist(g)
r_eq = cv2.equalizeHist(r)

# Merge the equalized color channels back into a single image
img_eq = cv2.merge((b_eq, g_eq, r_eq))

# Plot the original image and the enhanced image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_eq)
plt.title('Enhanced Image')

plt.show()
```

### Avoid the image being too bright

```python
# Load the image
img = cv2.imread('JGS_1.jpeg')

# Convert the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Define a sharpening kernel
kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

# Apply the sharpening kernel to the grayscale image
sharpened = cv2.filter2D(gray, -1, kernel)

# Convert the sharpened grayscale image back to color
img_sharp = cv2.cvtColor(sharpened, cv2.COLOR_GRAY2RGB)

# Plot the original image and the enhanced image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_sharp)
plt.title('Sharpened Image')

plt.show()
```

#### Return color

```python
# Load the image
img = cv2.imread('JGS_2.jpg')

# Convert the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split the image into its 3 color channels
b, g, r = cv2.split(img)

# Define a sharpening kernel
kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

# Apply the sharpening kernel to each color channel
b_sharp = cv2.filter2D(b, -1, kernel)
g_sharp = cv2.filter2D(g, -1, kernel)
r_sharp = cv2.filter2D(r, -1, kernel)

# Merge the sharpened color channels back into a single image
img_sharp = cv2.merge((b_sharp, g_sharp, r_sharp))

# Plot the original image and the enhanced image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_sharp)
plt.title('Sharpened Image')

plt.show()
```
